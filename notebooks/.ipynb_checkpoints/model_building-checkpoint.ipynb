Build a regression model.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/combined_data.csv')
```

```python
df
```

```python
# Splitting data into features (X) and target (y)

X = df[['free_bikes', 'latitude_x', 'latitude_y']]
y = df['rating']
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)

# Predictions
y_pred_linear = linear_model.predict(X)

# Evaluating model
mse = mean_squared_error(y, y_pred_linear)
print("Mean Squared Error:", mse)
```

Provide model output and an interpretation of the results. 

```python
import statsmodels.api as sm

X = sm.add_constant(X)  # Adding a constant column for the intercept
model = sm.OLS(y, X).fit()
print(model.summary())
```

**Notes**:
- From this model, we can see a very low R-Squared value. Meaning this is not a good model for predicting the rating of the restaurant based on its location and the number of free bikes at the nearby station.
- Both the variable have a p-value much greater than 0.05. This means that including both these variables make the model weak and does not the rating of the restaurant.
